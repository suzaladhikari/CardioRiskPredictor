```python
import pandas as pd 
import numpy as np 
```

```python
### Since the cleaned.csv has alot of biasness in the data, we will fix that using the random sampling 
data = pd.read_csv('../dataCleaning/cleaned.csv')
data.head(5)
```

```python
##Lets see the total counts of people with Heart Disease and without Heart Disease 
print(data['Heart_Disease'].value_counts(normalize=True) *100)
##As seen there are 91% of total No cases whereas only 8.08% of the Yes cases, which causes Biasness


```

```python
onlyYes = data[data['Heart_Disease'] == 'Yes']
onlyNo = data[data['Heart_Disease'] == 'No']
sampleOne = onlyNo.sample(n = 80000, replace = True)
sampleTwo = onlyYes.sample(n = 80000, replace = True)
merged = pd.concat([sampleOne,sampleTwo])
onlyCategorical = merged.select_dtypes('object')
print(onlyCategorical.columns)

for every in onlyCategorical.columns:
    print(f"The normalize value of {every} column is {onlyCategorical[every].value_counts(normalize=True)}")
```

### Since we have achieved equality in the distribution of the data, we can now change it to the csv file and prepare the model 

```python
merged.to_csv('forModel.csv')
```
